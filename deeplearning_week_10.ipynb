```python
#Image preprocessing library

from tensorflow.keras.preprocessing.image import load_img

#Keras Library to Convert image to array

from tensorflow.keras.preprocessing.image import img_to_array

#Transfer learning VGG of 16 layers
from keras.applications.vgg16 import VGG16

#Converts the image to fit the transfer learning model
from keras.applications.vgg16 import preprocess_input

#Decodes the prediction to string type
from keras.applications.vgg16 import decode_predictions
```

```python
#load sih tzu image and re-size it to 224x224
path = '/content/Transfer-Learning/Dogs_breed'
image = load_img(path+'/shih_tzu.jpg', target_size=(224, 224))

#Display image

import matplotlib.pyplot as plt

plt.imshow(image)
```

```python
# convert the dog image to numpy array

image = img_to_array(image)

image.shape
```

```python
# reshape data for the model

image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))

#prepare the image for the VGG model

image = preprocess_input(image)

#load the model

vgg = VGG16()
```

```python
# predict the probability across all output labels

prediction = vgg.predict(image)

```

```python
# convert the probabilities output to class labels

label = decode_predictions (prediction)

# retrieve the most likely result, e.g. highest probability
label = label [0] [0]

print(f'{label[1]} {round (label[2]*100,2)}%')
```
